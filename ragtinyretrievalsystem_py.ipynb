```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
model = SentenceTransformer('all-MiniLM-L6-v2')
notes = ['Fastapi builds backend apis',
          'python is a programming language ',
         'Exercise improves health',
         'SQLite stores data in local databases'
         ]
note_embedding = model.encode(notes)
question = ('How do i create apis ?')
question_embedding = model.encode([question])
similarity = cosine_similarity(question_embedding,note_embedding)
best_match_index = similarity.argmax()
best_note = notes[best_match_index]
print('most relevant note :')
print(best_note)
```
